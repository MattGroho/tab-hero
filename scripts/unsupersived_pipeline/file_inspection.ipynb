```python
from sklearn.cluster import KMeans
```

```python
difficulty_clustering_columns = [
    # 'content_hash',
    # 'difficulty_id',
    # 'instrument_id',
    'duration_sec',
    # 'rms_energy_mean',
    # 'rms_energy_std',
    # 'amplitude_envelope_min',
    # 'amplitude_envelope_max',
    # 'amplitude_envelope_range',
    'tempo_bpm',
    'n_notes',
    'notes_per_second_mean',
    'inter_note_ms_mean',
    'inter_note_ms_std',
    'inter_note_ms_median',
    'inter_note_ms_p25',
    'inter_note_ms_p75',
    'sustain_ratio',
    'chord_ratio',
    'fret_entropy',
    'hopo_ratio',
    'star_power_ratio',
    'tap_ratio',
    # 'mel_rms_mean',
    # 'mel_rms_std',
    # 'spectral_centroid_mean'
]


X = sample_100.loc[:, difficulty_clustering_columns]
X
```

```python
kmeans_diff = KMeans(
    n_clusters=5,
    random_state=42
).fit(X)

kmeans_diff.cluster_centers_
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

n_components = 10

pca = PCA(n_components=n_components).fit(X)

var_ratios = pca.explained_variance_ratio_
var_ratios_cum = np.tri(10) @ var_ratios
labels = [f"pc{i}" for i in range(1,n_components+1)]

# plt.bar(labels, var_ratios)

plt.bar(labels, var_ratios_cum)
for eigenval in pca.explained_variance_:
    print(eigenval)
```

```python
X_pca = pca.components_
kmeans_diff = KMeans(
    n_clusters=5,
    random_state=42
).fit(X_pca)

X_pca_pca = PCA(n_components=2).fit(kmeans_diff.cluster_centers_)
X_pca_pca.components_.
```
